```python
import pandas as pd
import json

data = []
with open('X-api/space-pics.json') as f:
    for line in f:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError as e:
            print(f"Skipping line due to error: {e}")

# parse out data and meta data from json file
df_data = pd.json_normalize(data, record_path=['data'])
df_meta = pd.json_normalize(data).drop('data', axis=1)

# Concatenate the DataFrames
df = pd.concat([df_data, df_meta], axis=1)
df.head()
```
